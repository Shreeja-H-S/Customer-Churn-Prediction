# Customer Churn Prediction - Data Cleaning

## Objective

The objective of this notebook is to clean and prepare the customer churn dataset for exploratory data analysis and machine learning.

### Steps

- Import Libraries
- Load Dataset
- Explore Dataset
- Check Missing Values
- Check Duplicates
- Check Data Types
- Generate Summary Statistics
- Save Cleaned Dataset

Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

Load the Dataset

```python
df = pd.read_csv("../data/raw/retail_customer_churn.csv")
```

Display the First 5 Rows

```python
df.head()
```

Check Dataset Shape

```python
print("Rows and Columns:", df.shape)
```

Display Column Names

```python
print(df.columns.tolist())
```

Check Dataset Information

```python
df.info()
```

Summary Statistics

```python
df.describe()
```

Check Missing Values

```python
df.isnull().sum()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

Check Duplicate Rows

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()
```

```python
df.dtypes
```

```python
df["churn_flag"].value_counts()
```

```python
df["churn_flag"].value_counts().plot(kind="bar")
plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Count")
plt.show()
```

```python
os.makedirs("../data/processed", exist_ok=True)
```

```python
df.to_csv("../data/processed/cleaned_customer_churn.csv", index=False)
```
